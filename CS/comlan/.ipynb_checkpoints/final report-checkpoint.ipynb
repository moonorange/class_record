```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
import math
import seaborn

def calc_sqrdiff(avec, bvec,xvec,yvec):
    diffmat = numpy.zeros((len(avec),len(bvec)))
    min = 1000
    mininda = 0
    minindb = 0
    for aind in range(len(avec)):
        for bind in range(len(bvec)):
            sqrdiff = numpy.square(yvec - (xvec*bvec[bind]+avec[aind])).sum()
            diffmat[aind][bind] = sqrdiff
            if sqrdiff < min:
                min = sqrdiff
                mininda = aind
                minindb = bind
    return diffmat,mininda,minindb,min
            

mat = np.hstack((pop,over65))
xvec = mat[:,0]
yvec = mat[:,1]
plt.scatter(xvec,yvec)
estb = numpy.corrcoef(xvec,yvec)[0][1]*numpy.std(yvec)/numpy.std(xvec)
esta =  numpy.mean(yvec)-estb*numpy.mean(xvec)
print(esta,estb) #xの係数　α=0.08963907803148624, β＝517362.78417243576
xmm = numpy.array((0,3.8*(10**6)))
ymm = xmm*estb+esta
plt.plot(xmm,ymm,color="red")
plt.show()
step = 0.01
amin = 0
amax = 2
bmin = 0
bmax = 1
avec = numpy.arange(amin,amax,step)
bvec = numpy.arange(bmin,bmax,step) 
   

diffmat,mininda,minindb,minsqdiff = calc_sqrdiff(avec,bvec,xvec,yvec)



```

```python
    
def k_means(num_class,data):
        d_size, n_features = data.shape
        centroids = np.loadtxt("FR3-initcenters.csv",delimiter=",")
        new_centroids = np.zeros((num_class, n_features))
        cluster = np.zeros(d_size)
        max_iter = 100

        for epoch in range(max_iter):

            for i in range(d_size):
                distances = np.sum((centroids - data[i]) ** 2, axis=1)
                cluster[i] = np.argsort(distances)[0]
                

            for j in range(num_class):
                new_centroids[j] = data[cluster==j].mean(axis=0)

            if np.sum(new_centroids == centroids) ==num_class:
                break
            centroids = new_centroids

        return cluster

```
