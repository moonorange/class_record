```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ts0 = np.loadtxt("week7ar0.csv",delimiter=",")
ts7 = np.loadtxt("week7ar7.csv",delimiter=",")

plt.plot(ts0)
plt.show()
lag = 1
plt.scatter(ts0[0:tslen-lag],ts0[lag:tslen])
plt.show()

rho1 = np.corrcoef(ts0[0:tslen-lag],ts0[lag:tslen])[0][1]
print(rho1)

plt.plot(ts7)
plt.show()
lag = 1
plt.scatter(ts7[0:tslen-lag],ts7[lag:tslen])
plt.show()
rho1 = np.corrcoef(ts7[0:tslen-lag],ts7[lag:tslen])[0][1]
print(rho1)

lag = 1
tslen = len(ts7)
xvec = ts7[0:tslen-lag]
yvec = ts7[lag:tslen]
xdevvec = xvec-xvec.mean()
ydevvec = yvec-yvec.mean()
cov = xdevvec.dot(ydevvec)/(tslen-1)
rho1 =cov/(xdevvec.std()*ydevvec.std())
rho1ver2 = np.corrcoef(ts7[0:tslen-lag],ts7[lag:tslen])[0][1]
print(rho1,rho1ver2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
tslen = 100
cor = 1.0
m = 0.2
s = 1.0
ts = np.zeros(tslen)
for i in range(1,tslen):
    rnd = (np.random.randn()+m)*s
    ts[i] = ts[i-1]*cor + rnd
plt.plot(ts)
plt.show()
tsdiff = ts[0:tslen-1]-ts[1:tslen]

plt.plot(tsdiff)
plt.show()

print(np.corrcoef(tsdiff[0:tslen-2],tsdiff[1:tslen-1]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
tslen = 100
cor = 0.
cor2 = 0.85
m = 0.0
s = 0.1
p = 4
ts = np.zeros(tslen)
ts[0] = 0
ts[1] = 0.05
ts[2] = 0.2
ts[3] = 0.1 
for i in range(p,tslen):
    rnd = (np.random.randn()+m)*s
    ts[i] = ts[i-1]*cor+ts[i-p]*cor2 + rnd
plt.plot(ts)
plt.show()
tsdiff = ts[0:tslen-p]-ts[p:tslen]

plt.plot(tsdiff)
plt.show()

print(np.corrcoef(tsdiff[0:tslen-p-1],tsdiff[1:tslen-p]))
```
