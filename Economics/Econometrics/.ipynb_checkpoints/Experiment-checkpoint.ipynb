```python
#　日本産が0、それ以外は1
def model_ja(x):
    if 'Honda' in x:
        r = 0
    elif "Mazda" in x:
        r = 0
    elif "Hyundai" in x:
        r = 0
    elif "Mitsubishi" in x:
        r = 0
    elif "Nissan" in x:
        r = 0
    elif "Subaru" in x:
        r = 0  
    elif "Toyota" in x:
        r = 0
    else:
        r = 1
    return r
```

```python
#　高級車が１、それ以外は０
def exp_brand(x):
    if 'Cadillac' in x:
        r = 1
    elif "Genesis" in x:
        r = 1
    elif "Alfa" in x:
        r = 1
    elif "BMW" in x:
        r = 1
    elif "Infiniti" in x:
        r = 1
    elif "Jaguar" in x:
        r = 1
    elif "Genesis" in x:
        r = 1
    elif "Land" in x:
        r = 1
    elif "Lincoin" in x:
        r = 1
    elif "Maserati" in x:
        r = 1
    elif "Mercedes-Benz" in x:
        r = 1
    elif "Porsche" in x:
        r = 1
    elif "Volvo" in x:
        r = 1
    elif "Audi" in x:
        r = 1
    elif "Lexus" in x:
        r = 1
    else:
        r = 0
    return r
```

```python
import sys
import matplotlib.pyplot as plt
from matplotlib import animation as ani
import numpy as np
import pandas as pd
import scipy.stats as st
from scipy.special import ndtri
```

```python
# 家賃

# 家賃を値段の順番に並び替え
price_ordered = np.sort(y_res1)

# 標準正規分布の逆関数(xの定義域と粒度は0-1の間をデータサイズの数分割したもの)
inv = ndtri(np.linspace(0, 1, data_size))#float(i)/len(price)) for i in range(len(price))] 

plt.title("Q-Q Plot", size=13)
plt.xlabel("Theoretical Quantailes")
plt.ylabel("Price")
plt.ylim(-0.42, 0.4)
plt.xlim(-3, 3)
plt.scatter(inv, price_ordered)
plt.show()
```
