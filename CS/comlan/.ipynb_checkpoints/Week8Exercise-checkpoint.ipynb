```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def calc_dist(datum, center):
    return np.square(datum-center).sum()

def reassign(numcluster,data,centers):
    assignment = dict()
    for idx  in range(len(data)):
        distlist_centers_datum = np.array([calc_dist(data[idx],centers[clidx]) for clidx in range(numcluster)])
        assignment[idx] = distlist_centers_datum.argmin()
    return assignment

def recalc_centers(numcluster,data,assignment):
    dim = np.shape(data)[1]
    centers = np.empty((numcluster,dim))
    for cid in range(numcluster):
        keylist = [k for k,v in assignment.items() if v == cid]
        centers[cid] = data[keylist].mean(axis=0)   
    return centers

def kmeans(numcluster, data,centers):
    dim = np.shape(data)[1]
    assignment = dict(zip(range(len(data)),[0]*len(data)))
    while(True):
        new_assignment = reassign(numcluster,data,centers)
        if new_assignment == assignment:
            break
        assignment = new_assignment
        centers = recalc_centers(numcluster,data,assignment)
    return assignment,centers

if __name__ == '__main__':
    numcluster = 3
    data = np.loadtxt("week8cluster.csv",delimiter=",")
    init_centers = np.array([[0.0,5.0], [5.0 , 0.0], [5.0, 10.0]])
    assignment,centers = kmeans(numcluster,data,init_centers)
    print(assignment)
    print(centers)
    
    for clidx in range(numcluster):
        clidxlist = [k for k,v in assignment.items() if v == clidx]
        cldata = data[clidxlist,:]
        plt.scatter(cldata[:,0],cldata[:,1])
        plt.scatter(centers[clidx,0],centers[clidx,1],marker="x",color="red",s=200)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def calc_dist(datum, center):
    return np.square(datum-center).sum()

def reassign(numcluster,data,centers):
    assignment = dict()
    for idx  in range(len(data)):
        distlist_centers_datum = np.array([calc_dist(data[idx],centers[clidx]) for clidx in range(numcluster)])
        assignment[idx] = distlist_centers_datum.argmin()
    return assignment

def recalc_centers(numcluster,data,assignment):
    dim = np.shape(data)[1]
    centers = np.empty((numcluster,dim))
    for cid in range(numcluster):
        keylist = [k for k,v in assignment.items() if v == cid]
        centers[cid] = data[keylist].mean(axis=0)   
    return centers

#EX8-2
def calc_distall(numcluster, data,centers,assignment):
    dist = 0
    for cid in range(numcluster):
        dist += np.square(data[[k for k,v in assignment.items() if v == cid],:]-centers[cid]).sum()
    return dist

#EX8-3
def plot_cluster(numcluster, data,centers,assignment,col):
        for clidx in range(numcluster):
            clidxlist = [k for k,v in assignment.items() if v == clidx]
            cldata = data[clidxlist,:]
            plt.scatter(cldata[:,0],cldata[:,1])
            plt.scatter(centers[clidx,0],centers[clidx,1],marker="x",color=col,s=200)
        plt.show()
        print(centers)
        print('J : {:.2f}'.format(calc_distall(numcluster,data,centers,assignment)))

def kmeans(numcluster, data,centers,withplot=1):
    assignment = dict(zip(range(len(data)),[0]*len(data)))
    
    while(True):
        if withplot:
            plot_cluster(numcluster, data,centers,assignment,'red')
        new_assignment = reassign(numcluster,data,centers)
        if withplot:
            plot_cluster(numcluster, data,centers,new_assignment,'black')
        if new_assignment == assignment:
            break
        assignment = new_assignment
        centers = recalc_centers(numcluster,data,assignment)
    J = calc_distall(numcluster,data,centers,assignment)
  #         print(centers)
    return assignment,centers,J

if __name__ == '__main__':
    data = np.loadtxt("week8cluster.csv",delimiter=",")
    dim = np.shape(data)[1]
    np.set_printoptions(precision=2, floatmode='maxprec_equal',suppress=True)

    init_centers = np.array([[0.0,5.0], [5.0 , 0.0], [5.0, 10.0]])
    numcluster = init_centers.shape[0]
    assignment,centers,J = kmeans(numcluster,data,init_centers,withplot=1)
#    print(assignment)
#    print(centers)
    plot_cluster(numcluster, data,centers,assignment,'red')
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def calc_dist(datum, center):
    return np.square(datum-center).sum()

def reassign(numcluster,data,centers):
    assignment = dict()
    for idx  in range(len(data)):
        distlist_centers_datum = np.array([calc_dist(data[idx],centers[clidx]) for clidx in range(numcluster)])
        assignment[idx] = distlist_centers_datum.argmin()
    return assignment

# EX8-6
def recalc_centers(numcluster,data,assignment,prev_centers):
    centers = np.empty((numcluster,dim))
    for cid in range(numcluster):
        keylist = [k for k,v in assignment.items() if v == cid]
# EX8-6
        if(keylist):
            centers[cid] = data[keylist].mean(axis=0) 
        else:
            centers[cid] = prev_centers[cid]
    return centers


def calc_distall(numcluster, data,centers,assignment):
    dist = 0
    for cid in range(numcluster):
        dist += np.square(data[[k for k,v in assignment.items() if v == cid],:]-centers[cid]).sum()
    return dist


def plot_cluster(numcluster, data,centers,assignment,col):
        for clidx in range(numcluster):
            clidxlist = [k for k,v in assignment.items() if v == clidx]
            cldata = data[clidxlist,:]
            plt.scatter(cldata[:,0],cldata[:,1])
            plt.scatter(centers[clidx,0],centers[clidx,1],marker="x",color=col,s=200)
        plt.show()
        print(centers)
        print('J : {:.2f}'.format(calc_distall(numcluster,data,centers,assignment)))

def kmeans(numcluster, data,centers,withplot=1,max_iter=1000):
    assignment = dict(zip(range(len(data)),[0]*len(data)))
    
# EX8-6
    for i in range(max_iter):
        if withplot:
            plot_cluster(numcluster, data,centers,assignment,'red')
        new_assignment = reassign(numcluster,data,centers)
        if withplot:
            plot_cluster(numcluster, data,centers,new_assignment,'black')
        if new_assignment == assignment:
            break
        assignment = new_assignment
        centers = recalc_centers(numcluster,data,assignment,centers)
    J = calc_distall(numcluster,data,centers,assignment)
  #         print(centers)
    return assignment,centers,J

if __name__ == '__main__':
    data = np.loadtxt("week8cluster.csv",delimiter=",")
    dim = np.shape(data)[1]
    np.set_printoptions(precision=2, floatmode='maxprec_equal',suppress=True)

    init_centers = np.array([[0.0,5.0], [5.0 , 0.0], [20.0, 10.0]])
    numcluster = init_centers.shape[0]
    assignment,centers,J = kmeans(numcluster,data,init_centers,withplot=1,max_iter=100)
#    print(assignment)
#    print(centers)
    plot_cluster(numcluster, data,centers,assignment,'red')
```
