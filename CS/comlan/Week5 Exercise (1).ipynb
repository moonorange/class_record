```python
import matplotlib.pyplot as plt
import numpy
%matplotlib inline


cdic = dict()
for c in range(ord("a"),ord("z")+1):
    cdic[chr(c)] = 0
print(cdic.keys)

f = open("icu_web_text.txt")
wdic =  dict()
for line in f:
    trimed = line.rstrip()
    words = trimed.split()
    for w in words:
        if w in wdic:
            wdic[w] +=1
        else:
            wdic[w] = 1
        
        for c in w:
            cdic[c]+=1    
f.close()
str_freqlist = list(cdic.values())


wclist = numpy.array(list((wdic.values())))
print(wclist.max())
wchist = numpy.histogram(wclist, bins=wclist.max())
x = range(1,wclist.max()+1)
plt.plot(x,wchist[0],marker="o")
plt.xlabel("number of appearance")
plt.ylabel("number of words")
plt.xscale("log")
plt.yscale("log")
plt.show()
wdic_sorted = sorted(wdic.items(), key=lambda x: x[1], reverse=True)
#print(wdic_sorted)


gcdic = dict()
f = open("english_letter_freq.txt")
for line in f:
    trimed = line.rstrip()
    [c,f] = trimed.split()
    gcdic[c] = float(f)

alp_freqlist = list(gcdic.values())

# cosine similarity
alp_freqvec = numpy.array(alp_freqlist)
str_freqvec = numpy.array(str_freqlist)
alp_norm = numpy.sqrt(numpy.square(alp_freqvec).sum())
str_norm = numpy.sqrt(numpy.square(str_freqvec).sum())
cossim = alp_freqvec.dot(str_freqvec)/(alp_norm*str_norm)
print(cossim)

# end

plt.plot(alp_freqvec)
plt.plot(str_freqvec/str_freqvec.sum(),color="red")
plt.xlabel("alphabet")
plt.ylabel("number of appearance")
plt.show()
plt.scatter(alp_freqvec,str_freqvec)
numpy.corrcoef(alp_freqvec,str_freqvec)
```
