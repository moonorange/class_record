```python
#%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as numpy

fmpop_perage = numpy.loadtxt("./nenreibetu_jinkouhoukokusyo30.csv",delimiter=",",dtype=int)
print(fmpop_perage.shape)
age = fmpop_perage[:,0] # pick 1st column
male = fmpop_perage[:,1] # pick 2st column
female = fmpop_perage[:,2] # pick 3st column
plt.scatter(male,female)
plt.show()

all = male + female  # vector element-wise sum
maleratio = male/all  # vector element-wise division
plt.plot(age,maleratio)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy 
import seaborn


month_temprature = numpy.loadtxt("./tokyo_monthly_average_max_2008_2017.csv",delimiter=",",dtype=str) # 1st column include str
month_average = month_temprature[:,1].astype('float64') # pick 2nd column and covert it to float
month_num = numpy.arange(1,121,1) # prepare the x-axis value
print(month_average.shape, month_num.shape)

plt.plot(month_num, month_average)
plt.show()

mamat = month_average.reshape(10,12)
seaborn.heatmap(mamat,cmap="coolwarm")
plt.show()

print(mamat.mean(axis=0)[7])  #calc mean for column axis and take 8-th value
print(mamat[:,7].mean())    #pick 8-th column and calc mean

plt.scatter(mamat[:,5],mamat[:,7])
plt.show()
seaborn.heatmap(numpy.corrcoef(mamat.T))
```
